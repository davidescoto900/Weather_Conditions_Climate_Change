## Data Wrangling

```python
# Create a list of the columns containing "Oslo" in their names
oslo_list = list(x for x in X.columns if x.find('OSLO') >=0)
oslo_list
```

```python
# Create a dataframe with those columns
df_oslo = X[oslo_list]
df_oslo
```

```python
# Reduce answers dataset to Oslo answers only
answers_oslo = answers['OSLO_pleasant_weather']
answers_oslo
```

```python
df_oslo.shape # observations dataset has the correct shape
```

```python
answers_oslo.shape # predictions dataset has the correct shape
```

## Reshaping the model

```python
X2 = df_oslo
```

```python
y2 = answers_oslo
```

```python
# Turn X2 and y2 from df to arrays
X = np.array(X2)
y = np.array(y2)
```

```python
X.shape
```

```python
y.shape
```

## Data Split

```python
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 42)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

## Random Forest Model

```python
# Create a RF classifier
clf = RandomForestClassifier(n_estimators = 100)#, max_depth=5)  
  
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
clf.fit(X_train, y_train)
```

```python
# Perform predictions on the test dataset
y_pred = clf.predict(X_test)
  
# Use metrics module for accuracy calculation
print("Model Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

```python
# Class-names = {0:'Unpleasant Weather', 1:'Pleasant Weather'}

fig = plt.figure(figsize=(80,40))
plot_tree(clf.estimators_[6], fontsize = 20, filled=True);
```

```python
fig.savefig(os.path.join(path, 'random_forest_Oslo.png'),bbox_inches='tight')
```

## Uncovering Feature Importance

```python
# Retrieve feature importances from the trained model
newarray = clf.feature_importances_
print(clf.feature_importances_.shape)
newarray
```

```python
# Create a list of weather features
wx_list = [feature.replace('OSLO_', '') for feature in oslo_list]
wx_list
```

```python
important = pd.Series(newarray, index = wx_list)
important
```

```python
# Plot the results

%matplotlib inline

plt.style.use('fivethirtyeight')
# list of x locations for plotting
x_values = list(range(len(newarray)))
print(wx_list)

plt.bar(x_values, newarray, orientation = 'vertical')
plt.xticks(x_values, wx_list, rotation='vertical')
plt.ylabel('Importance'); plt.xlabel('Feature'); plt.title('Features Importances for Oslo (all years)');

plt.savefig(os.path.join(path, 'Oslo_feature_importances.png'), bbox_inches='tight')
plt.show()
```
